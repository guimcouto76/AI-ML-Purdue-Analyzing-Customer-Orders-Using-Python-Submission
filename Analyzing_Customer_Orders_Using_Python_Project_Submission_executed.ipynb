## 1.3 Use a dictionary where keys are customer names and values are lists of ordered products

```python
customer_product_dictionary = {}

# setdefault avoids a "key not in dict" check each iteration
for customer, product, price, category in customer_orders:
    customer_product_dictionary.setdefault(customer, []).append(product)

for customer, products in customer_product_dictionary.items():
    print(f"{customer}: {products}")
```

# 2. Classify products by category

# 3. Analyze customer orders

## 3.1 Use a loop to calculate the total amount each customer spends

```python
customer_spending = {}

for customer, product, price, category in customer_orders:
    customer_spending[customer] = customer_spending.get(customer, 0) + price

for customer, total in customer_spending.items():
    print(f"{customer}: ${total:.2f}")
```

## 3.2 Classify customers based on total purchase value

- Above $100: High-value buyer
- Between $50 and $100: Moderate buyer
- Below $50: Low-value buyer

I read "between $50 and $100" as inclusive on both ends, so the rules become `total > 100`, `50 <= total <= 100`, and `total < 50`.

```python
customer_classification = {}

# > 100 high, [50, 100] moderate, < 50 low
for customer, total in customer_spending.items():
    if total > 100:
        classification = "High-value buyer"
    elif 50 <= total <= 100:
        classification = "Moderate buyer"
    else:
        classification = "Low-value buyer"
    customer_classification[customer] = classification

for customer, classification in customer_classification.items():
    print(f"{customer}: {classification}")
```

### 3. Response

Seven of the ten customers came out as high-value buyers. That makes sense once you look at the totals: a single Electronics purchase like a laptop or a smartphone is enough to push a customer past $100. The remaining three (Brian, Jack, Daniel) only bought cheaper Clothing, Accessories, and Office Supplies items. This split is useful for marketing because it lets us send different offers to each segment.

# 4. Generate business insights

## 4.1 Calculate the total revenue per product category and store it in a dictionary

```python
category_revenue = {}

for customer, product, price, category in customer_orders:
    category_revenue[category] = category_revenue.get(category, 0) + price

for category, revenue in sorted(category_revenue.items(), key=lambda x: x[1], reverse=True):
    print(f"{category}: ${revenue:.2f}")
```

## 4.2 Extract unique products from all orders using a set

```python
unique_products = {product for customer, product, price, category in customer_orders}

for product in sorted(unique_products):
    print("-", product)
```

## 4.3 Use a list comprehension to find all customers who purchased electronics

```python
# list comprehension first, then dedup with set() and sort
electronics_customers_all = [
    customer
    for customer, product, price, category in customer_orders
    if category == "Electronics"
]

electronics_customers = sorted(set(electronics_customers_all))

print("Raw list comprehension result (may contain duplicates):")
print(electronics_customers_all)

print("\nUnique customers who purchased Electronics:")
for customer in electronics_customers:
    print("-", customer)
```

## 4.4 Identify the top three highest-spending customers using sorting

```python
top_three_customers = sorted(
    customer_spending.items(),
    key=lambda item: item[1],
    reverse=True
)[:3]

for rank, (customer, total) in enumerate(top_three_customers, start=1):
    print(f"{rank}. {customer}: ${total:.2f}")
```

## 4.5 Extract the high-value customers and most frequently purchased products

```python
high_value_customers = [
    customer
    for customer, classification in customer_classification.items()
    if classification == "High-value buyer"
]

print("High-Value Customers:")
for customer in high_value_customers:
    print("-", customer)

product_frequency = {}
for customer, product, price, category in customer_orders:
    product_frequency[product] = product_frequency.get(product, 0) + 1

sorted_frequency = sorted(
    product_frequency.items(),
    key=lambda item: (-item[1], item[0])
)

print("\nMost Frequently Purchased Products (top 10):")
print(f"{'Product':<20}{'Times Purchased':>16}")
print("-" * 36)
for product, count in sorted_frequency[:10]:
    print(f"{product:<20}{count:>16}")

repeat_products = [(p, c) for p, c in sorted_frequency if c > 1]
print(f"\nProducts purchased by more than one customer: {len(repeat_products)}")
print(f"Products purchased only once: {len(sorted_frequency) - len(repeat_products)}")
```

### 4. Response

Electronics is the strongest category by revenue, mostly because of high-priced items like laptops, smartphones, and tablets. Home Essentials comes in second thanks to vacuums, microwaves, and coffee makers. Sorting the spending dictionary and slicing the first three entries is a clean way to get the top spenders without tracking maxes manually.

## 5.1 Print a summary of each customer's total spending and their classification

## 5.2 Use set operations to find customers who purchased from multiple categories

```python
customer_categories = {}

for customer, product, price, category in customer_orders:
    customer_categories.setdefault(customer, set()).add(category)

multiple_category_customers = {
    customer
    for customer, categories in customer_categories.items()
    if len(categories) > 1
}

for customer in sorted(multiple_category_customers):
    print(f"{customer}: {customer_categories[customer]}")
```

## 5.3 Identify common customers who bought both electronics and clothing

```python
electronics_customer_set = {
    customer
    for customer, product, price, category in customer_orders
    if category == "Electronics"
}

clothing_customer_set = {
    customer
    for customer, product, price, category in customer_orders
    if category == "Clothing"
}

electronics_and_clothing_customers = electronics_customer_set.intersection(clothing_customer_set)

print("Customers Who Bought Both Electronics and Clothing:")
for customer in sorted(electronics_and_clothing_customers):
    print("-", customer)

print("\nElectronics Customers:", electronics_customer_set)
print("Clothing Customers:", clothing_customer_set)
print("Intersection:", electronics_and_clothing_customers)
```

## 5.4 Retrieve and modify customer records using dictionary methods

```python
# work on a copy so the demo does not mutate the dict used in the final report
demo_products = {
    customer: list(products)
    for customer, products in customer_product_dictionary.items()
}

selected_customer = "Alice Johnson"

alice_products = demo_products.get(selected_customer, [])
print(f"Original products for {selected_customer}: {alice_products}")

demo_products.setdefault(selected_customer, []).append("USB-C Cable")
print(f"Updated products for {selected_customer} (demo copy): {demo_products[selected_customer]}")

print(f"\nOriginal dictionary unchanged for {selected_customer}: "
      f"{customer_product_dictionary[selected_customer]}")

new_customer = "Karen Patel"
demo_products.setdefault(new_customer, []).append("Gift Card")
print(f"Added new customer record in demo copy -> {new_customer}: {demo_products[new_customer]}")
```

## 5.5 Use set operations to find common and unique products across categories

The brief asks for set operations across categories. Since each product belongs to exactly one category, the intersection of any two category-product sets is empty, which is what the output below shows. The same cell also prints the products unique to each category and the union of products across Electronics, Clothing, and Home Essentials.

```python
products_by_category = {}
for customer, product, price, category in customer_orders:
    products_by_category.setdefault(category, set()).add(product)

print("Products per category (set per category):")
for category in sorted(products_by_category):
    print(f"  {category}: {sorted(products_by_category[category])}")

electronics_products = products_by_category.get("Electronics", set())
clothing_products    = products_by_category.get("Clothing", set())
home_products        = products_by_category.get("Home Essentials", set())

unique_to_electronics = electronics_products - clothing_products - home_products
unique_to_clothing    = clothing_products - electronics_products - home_products
electronics_and_clothing_products = electronics_products & clothing_products
all_three_categories_union = electronics_products | clothing_products | home_products

print("\nUnique to Electronics (set difference):")
print(sorted(unique_to_electronics))

print("\nUnique to Clothing (set difference):")
print(sorted(unique_to_clothing))

print("\nProducts that appear in BOTH Electronics and Clothing (intersection):")
print(electronics_and_clothing_products or "(empty - categories are disjoint by design)")

print("\nUnion of products in Electronics, Clothing and Home Essentials:")
print(f"  {len(all_three_categories_union)} unique products")
print(f"  {sorted(all_three_categories_union)}")
```

### 5. Response

Putting the totals, classifications, and product lists into one summary made the patterns easier to read. The set operations in 5.2 and 5.3 turned customer segmentation into one-line expressions: `len(categories) > 1` for multi-category buyers and `&` for the Electronics-and-Clothing overlap. In 5.4 I worked on a copy of the customer-to-products dictionary so the demo modifications would not affect the final report.

# Final Report and Business Recommendations

## Customer Classification Summary

```python
classification_summary = []

for customer, total in customer_spending.items():
    classification_summary.append({
        "Customer": customer,
        "Total Spending": total,
        "Classification": customer_classification[customer],
        "Number of Categories Purchased": len(customer_categories[customer])
    })

classification_summary = sorted(
    classification_summary,
    key=lambda item: item["Total Spending"],
    reverse=True
)

header = f"{'Rank':>4}  {'Customer':<18}{'Spending':>14}{'Classification':>22}{'#Cats':>8}"
print(header)
print("-" * len(header))

for rank, row in enumerate(classification_summary, start=1):
    amount = f"${row['Total Spending']:.2f}"
    print(
        f"{rank:>4}  "
        f"{row['Customer']:<18}"
        f"{amount:>14}"
        f"{row['Classification']:>22}"
        f"{row['Number of Categories Purchased']:>8}"
    )

print("-" * len(header))

total_revenue = sum(customer_spending.values())
high_count = sum(1 for r in classification_summary if r["Classification"] == "High-value buyer")
mod_count  = sum(1 for r in classification_summary if r["Classification"] == "Moderate buyer")
low_count  = sum(1 for r in classification_summary if r["Classification"] == "Low-value buyer")

print(f"\nTotal customers:   {len(classification_summary)}")
print(f"Total revenue:     ${total_revenue:.2f}")
print(f"High-value buyers: {high_count}")
print(f"Moderate buyers:   {mod_count}")
print(f"Low-value buyers:  {low_count}")
```

## Final Business Insights

A few things stand out from the numbers and are worth passing along to the business team:

1. Electronics is the strongest category, accounting for about 75% of total revenue. It is the obvious area to focus inventory and promotions on.

2. Home Essentials is a solid second category. Items like vacuum cleaners, microwaves, and coffee makers add up quickly, so a seasonal promotion could increase the average order value.

3. Seven of the ten customers are high-value buyers and they generate almost all of the revenue. Loyalty perks or exclusive discounts would help retain them.

4. The two moderate buyers (Brian and Jack) spend close to $100. A small incentive such as free shipping above a threshold could move them into the high-value group.

5. The one low-value buyer (Daniel) only spent $40, all on cheaper items. A targeted coupon or a small recommendation email is a low-cost way to try to re-engage him.

6. Most customers shop across multiple categories already, so cross-category campaigns are likely to work well for this audience.

7. Five customers bought both Electronics and Clothing. They are a clear target for bundle offers such as smartwatch + running shoes or tablet + backpack.
